```
#Let's clean up all the zeros
import numpy as np
pat= pat.replace(0, np.nan)
pat = pat.fillna(method='ffill')
pat = pat.fillna(method='bfill')
```

```
#features engineering

pat["Monthly after tax income"]= pat["FINATXEM"]/12
pat["MPC"]=  pat['Total expenses']/12/pat['Monthly after tax income']
pat['Necessity_ratio']= ((pat['Total Housing']+ pat['Total Food']+pat['Total Transportation cost'])/3)/pat['Monthly after tax income']
pat['Luxury_ratio']=(pat["Luxury expenses"]/3)/pat['Monthly after tax income']
pat['Savings_ratio']=([pat['possible savings']])/pat['Monthly after tax income']

pat['delta_necessity']= 0.5 - pat
```
